# Trades features


Average of trades in last X time steps

```python
# first average trades per day and timestamp
test_trades_condensed = test_trades.groupby(['day','timestamp','symbol']).agg({
  'price': 'mean',
  'quantity': 'sum' 
}).reset_index()

# sort just in case
test_trades_condensed = test_trades_condensed.sort_values(by=['day', 'timestamp', 'symbol']).reset_index(drop=True)
test_trades_condensed
```

```python
x_periods = [3, 5, 7, 10]

def weighted_avg(prices, quantities):
    weighted_sum = sum(p * q for p, q in zip(prices, quantities))
    total_quantity = sum(quantities)
    
    return weighted_sum / total_quantity if total_quantity != 0 else 0

# Periods for rolling window
x_periods = [3, 5, 7, 10]

for period in x_periods:
    avg_prices = []
    
    for i in range(len(test_trades_condensed)):
        # For the first (period - 1) rows, append NaN to avg_prices
        if i < period - 1:
            avg_prices.append(None)
        else:
            start_index = max(0, i - period + 1)
            prices_window = test_trades_condensed['price'].iloc[start_index:i+1]
            quantities_window = test_trades_condensed['quantity'].iloc[start_index:i+1]
            
            # Compute weighted average for the current window
            avg_price = weighted_avg(prices_window, quantities_window)
            avg_prices.append(avg_price)
    
    # Add the computed average price to the DataFrame
    test_trades_condensed[f'avg_price_period_{period}'] = avg_prices
```

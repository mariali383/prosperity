```python
max_timestamp = prices2['timestamp'].max()

# Step 2: Shift the timestamps in prices1 so they continue after prices2
# Assume each day spans up to timestamp = 999900, so we shift by 1,000,000
# (or use max_timestamp + 100 to be exact if you want to generalize)
prices1_shifted = prices1.copy()
prices1_shifted['timestamp'] += (max_timestamp + 100)  # or 1000000

# Optional: If you want to reset the 'day' column too, to keep things clean
prices1_shifted['day'] = prices1_shifted['day'] + 1  # -1 becomes 0

# Step 3: Concatenate the dataframes
test_prices = pd.concat([prices2, prices1_shifted], ignore_index=True)

# Step 4 (optional): Sort by timestamp and product
test_prices = test_prices.sort_values(by=['timestamp', 'product']).reset_index(drop=True)

test_prices
```

```python
# Step 1: Pivot the dataframe on timestamp and day, with product as column level
flattened = test_prices.pivot(index=["timestamp", "day"], columns="product")

# Step 2: Flatten the multi-level column names
flattened.columns = [f"{product.lower()}_{col}" for col, product in flattened.columns]

# Step 3: Reset index to make it a flat dataframe again
flattened = flattened.reset_index()

test_prices = flattened
```

```python
max_timestamp = trades2['timestamp'].max()

# Step 2: Shift the timestamps in prices1 so they continue after prices2
# Assume each day spans up to timestamp = 999900, so we shift by 1,000,000
# (or use max_timestamp + 100 to be exact if you want to generalize)
trades1_shifted = trades1.copy()
trades1_shifted['timestamp'] += (1000000)  # or 1000000

# Step 3: Concatenate the dataframes
test_trades = pd.concat([trades2, trades1_shifted], ignore_index=True)

# Step 4 (optional): Sort by timestamp and product
test_trades = test_trades.sort_values(by=['timestamp', 'symbol']).reset_index(drop=True)

test_trades
```

```python
# Step 1: Compute price * quantity and keep total volume
test_trades['total_value'] = test_trades['price'] * test_trades['quantity']

# Step 2: Group by timestamp and symbol to sum value and volume
aggregated = (
    test_trades
    .groupby(['timestamp', 'symbol'])
    .agg(total_value=('total_value', 'sum'), total_volume=('quantity', 'sum'))
    .reset_index()
)

# Step 3: Pivot to wide format for each symbol
values_pivot = aggregated.pivot(index='timestamp', columns='symbol', values='total_value').rename(
    columns={
        'KELP': 'kelp_value',
        'RAINFOREST_RESIN': 'rainforest_resin_value',
        'SQUID_INK': 'squid_ink_value'
    }
)

volumes_pivot = aggregated.pivot(index='timestamp', columns='symbol', values='total_volume').rename(
    columns={
        'KELP': 'kelp_volume',
        'RAINFOREST_RESIN': 'rainforest_resin_volume',
        'SQUID_INK': 'squid_ink_volume'
    }
)

# Step 4: Merge with full timestamp range
full_timestamps = pd.DataFrame({'timestamp': np.arange(test_trades['timestamp'].min(), test_trades['timestamp'].max() + 100, 100)})
combined = full_timestamps.merge(values_pivot, on='timestamp', how='left').merge(volumes_pivot, on='timestamp', how='left')

rolling_vwap_df = combined[['timestamp']].copy()



# Start with just the timestamp column
rolling_vwap_df = combined[['timestamp']].copy()

# Compute rolling VWAPs
rolling_windows = [2, 3, 5, 7, 10, 20]
for rolling_window in rolling_windows:
    for symbol in ['kelp', 'rainforest_resin', 'squid_ink']:
        value_col = f'{symbol}_value'
        volume_col = f'{symbol}_volume'
        vwap_col = f'{symbol}_vwap_rolling_{rolling_window}'

        rolling_vwap_df[vwap_col] = (
            combined[value_col].rolling(window=rolling_window, min_periods=1).sum() /
            combined[volume_col].rolling(window=rolling_window, min_periods=1).sum()
        )


rolling_vwap_df.tail()
```

```python
vwap_cols = [col for col in test_prices.columns if 'vwap_rolling' in col]
test_prices = test_prices.drop(columns=vwap_cols)

# Now merge again
test_prices = test_prices.merge(rolling_vwap_df, on='timestamp', how='left')
```

```python
# Make sure all the VWAP columns are already merged into test_prices
# Now calculate the difference between VWAP and mid_price for each rolling window

symbols = ['squid_ink', 'kelp', 'rainforest_resin']
rolling_windows = [2, 3, 5, 7, 10, 20]

for symbol in symbols:
    mid_col = f"{symbol}_mid_price"
    for window in rolling_windows:
        vwap_col = f"{symbol}_vwap_rolling_{window}"
        diff_col = f"{symbol}_vwap_{window}_diff"
        test_prices[diff_col] = test_prices[vwap_col] - test_prices[mid_col]

```

# Creating Target Features
- Aim to look forward a given amount
- Aim to capture this first via absolute change
- Lookback periods potentially ranging from 1-10 timestamps ahead

```python
lookahead_periods = [1, 2, 3, 5, 7, 10]
target_cols = ['kelp_mid_price', 'rainforest_resin_mid_price', 'squid_ink_mid_price']

for col in target_cols:
    for period in lookahead_periods:
        new_col = f"{col}_fwd_diff_{period}"
        test_prices[new_col] = test_prices[col].shift(-period) - test_prices[col]
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

def linear_regression_plot(df, feature_col, target_col):
    """
    Performs linear regression to predict target_col using feature_col and plots the result.
    Also prints the correlation coefficient.
    """
    # Drop rows where either feature or target column contains NaN values
    df_clean = df.dropna(subset=[feature_col, target_col])
    
    # Prepare the data
    X = df_clean[[feature_col]].values  # Feature (2D for sklearn)
    y = df_clean[target_col].values     # Target (1D)

    # Compute correlation coefficient
    corr = np.corrcoef(df_clean[feature_col], df_clean[target_col])[0, 1]
    
    # Create and fit the linear regression model
    model = LinearRegression()
    model.fit(X, y)
    
    # Get slope and intercept
    gradient = model.coef_[0]
    intercept = model.intercept_
    
    # Predict values
    y_pred = model.predict(X)
    
    # Plotting
    plt.figure(figsize=(6, 6))
    plt.scatter(df_clean[feature_col], df_clean[target_col], color='blue', label='Data points')
    plt.plot(df_clean[feature_col], y_pred, color='red', label=f'Regression Line (y = {gradient:.2f}x + {intercept:.2f})')
    
    plt.title(f'Linear Regression: {target_col} vs {feature_col}')
    plt.xlabel(feature_col)
    plt.ylabel(target_col)
    plt.legend()
    plt.grid(True)
    plt.show()
    
    # Print results
    print(f"Gradient (Slope): {gradient:.2f}")
    print(f"Intercept: {intercept:.2f}")
    print(f"Correlation Coefficient: {corr:.4f}")

# Example usage:
linear_regression_plot(test_prices, 'squid_ink_mid_price_diff_ema_20', 'squid_ink_mid_price_fwd_diff_3')

linear_regression_plot(test_prices, 'kelp_vwap_2_diff', 'kelp_mid_price_fwd_diff_1')
linear_regression_plot(test_prices, 'kelp_vwap_3_diff', 'kelp_mid_price_fwd_diff_1')
linear_regression_plot(test_prices, 'kelp_vwap_5_diff', 'kelp_mid_price_fwd_diff_1')
linear_regression_plot(test_prices, 'kelp_vwap_7_diff', 'kelp_mid_price_fwd_diff_1')
linear_regression_plot(test_prices, 'kelp_vwap_10_diff', 'kelp_mid_price_fwd_diff_1')
linear_regression_plot(test_prices, 'kelp_vwap_20_diff', 'kelp_mid_price_fwd_diff_1')
linear_regression_plot(test_prices, 'squid_ink_vwap_2_diff', 'squid_ink_mid_price_fwd_diff_1')
linear_regression_plot(test_prices, 'squid_ink_vwap_3_diff', 'squid_ink_mid_price_fwd_diff_1')
linear_regression_plot(test_prices, 'squid_ink_vwap_5_diff', 'squid_ink_mid_price_fwd_diff_1')
linear_regression_plot(test_prices, 'squid_ink_vwap_7_diff', 'squid_ink_mid_price_fwd_diff_1')
linear_regression_plot(test_prices, 'squid_ink_vwap_10_diff', 'squid_ink_mid_price_fwd_diff_1')
linear_regression_plot(test_prices, 'squid_ink_vwap_20_diff', 'squid_ink_mid_price_fwd_diff_1')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import pandas as pd
def model_param_calc(dataframe, feature_columns, target_column, lower_bound, upper_bound):
    """
    Train and evaluate a linear regression model with customizable features and target column.

    Parameters:
    dataframe (pd.DataFrame): Input DataFrame containing the features and target.
    feature_columns (list): List of column names to be used as features.
    target_column (str): Name of the column to be used as the target.

    Returns:
    dict: A dictionary containing the model, intercept, coefficients, R-squared, and mean squared error.

    """

    # Drop rows with NaN values in the selected features and target
    filtered_df = dataframe.dropna(subset=feature_columns + [target_column])

    # Extract features (X) and target (y)
    X = filtered_df[feature_columns].values  # 2D array
    y = filtered_df[target_column].values  # 1D array

    # Train the linear regression model
    model = LinearRegression()
    model.fit(X, y)

    # Make predictions
    predicted_y = model.predict(X)

    # Create a mask for rows to keep (outside the range)
    mask = (predicted_y < lower_bound) | (predicted_y > upper_bound)

    X_filtered = X[mask]
    y_filtered = y[mask]
    predicted_y_filtered = predicted_y[mask]

    # Evaluate the model
    r2 = r2_score(y_filtered, predicted_y_filtered)
    mse = mean_squared_error(y_filtered, predicted_y_filtered)

    totalScore = 0

    posTotal, negTotal = 0, 0
    posHits, negHits = 0, 0
    posHitRate, negHitRate = 0, 0
    for i in range(len(predicted_y_filtered)):
        
        # If predicted show strong signal, check whether signal valid or not
        if predicted_y_filtered[i] > upper_bound:
            posTotal += 1
            if y_filtered[i] > 0:
                posHits += 1
            totalScore += y_filtered[i]

        if predicted_y_filtered[i] < lower_bound:
            negTotal += 1
            if y_filtered[i] < 0:
                negHits += 1
            totalScore -= y_filtered[i]

    avgScore = totalScore/(posTotal + negTotal)

    if posTotal > 0 and negTotal > 0:
        posHitRate = posHits/posTotal*100
        negHitRate = negHits/negTotal*100   

    frequency = posTotal + negTotal

    # Return the results
    return {
        "model": model,
        "features": feature_columns,
        "intercept": model.intercept_,
        "coefficients": model.coef_,
        "r_squared": r2,
        "mean_squared_error": mse,
        "pos_hit_rate": posHitRate,
        "neg_hit_rate": negHitRate,
        "total_score": totalScore,
        "avg_score": avgScore,
        "frequency": frequency
    }
```
